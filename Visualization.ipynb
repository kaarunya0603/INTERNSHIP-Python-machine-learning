```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"\Desktop\n\data-2.csv")
```

```python
df.head()

```

```python
df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"]); df.groupby(df["InvoiceDate"].dt.date)["Quantity"].sum().plot(figsize=(12,5))
```

```python
df.groupby("Country")["Quantity"].sum().sort_values(ascending=False).head(5).plot(kind="pie", autopct="%1.1f%%")
```
