```python
#Types of plot
 # Numerical Dataplot
   # 1.Distribution plot
   # 2.Join plot - Combines scatter and histogram (or KDE) to show bivariate relationships.
   # 3.Distribution plot -Plots pairwise relationships between multiple variables in a dataset.
   # 4.pair plot - A general function for drawing categorical plots like box, violin, strip, etc.
   # 5.categorical ploting -
   # 6.Barplot- Displays mean values of a categorical variable with confidence intervals.
   # 7.Count plot -Shows the count of observations for each category.
   # 8.Boxplot-Displays the distribution of a variable using quartiles and outliers.
   # 9.Violin Plot-Combines boxplot and KDE to show data distribution.
   # 10.Strip Plot-Shows all individual data points on a categorical axis with optional jittering.
   # 11.Factor Plot/ General Plot-Deprecated version of catplot for drawing categorical plots.
   # 12.Matrix Plot- Used to display matrix-style plots, often for correlations or comparisons.
   # 13.Heat Map-Displays data in matrix form with color intensity indicating values.
   # 15.Regression plot-Plots data and a linear regression model fit.
#TYPES OF GRID
 # 1.pairGrid-Creates a grid of subplots for plotting pairwise relationships between multiple variables.
 # 2.Faset Grid-Allows plotting the same type of plot across subsets of data (facets) in a grid layout.
```
