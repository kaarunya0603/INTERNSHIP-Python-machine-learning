```python
import plotly_express as px  # The plotly.express module (usually imported as px) contains functions 
                             # that can create entire figures at once, and is referred to as Plotly Express or PX.
```

```python
gapminder =px.data.gapminder() # load built-in gapminder dataset from plotly 
```

```python
gapminder2007 =gapminder.query("year==2007")
```

```python
px.scatter(gapminder2007, x="gdpPercap", y="lifeExp")
```

```python
px.scatter(gapminder2007, x="gdpPercap",y="lifeExp",color="continent") 
# This line creates a scatter plot using `px.scatter()` from the Plotly Express library, 
# showing the relationship between GDP per capita (`x="gdpPercap"`) 
# and life expectancy (`y="lifeExp"`) from the `gapminder2007` dataset, with points colored by continent (`color="continent"`).
```

```python
gapminder.head() # First 5 data info from dataset
```

```python
px.scatter(gapminder2007,x="gdpPercap",y="lifeExp",color="continent",size="pop",size_max=60)
# creates a scatter plot using Plotly Express (`px.scatter()`) with GDP per capita (`x="gdpPercap"`) on the x-axis, 
# life expectancy (`y="lifeExp"`) on the y-axis, points colored by continent (`color="continent"`), 
# and sized by population (`size="pop"`), where the maximum marker size is capped at 60 (`size_max=60`).
```

```python
px.scatter(gapminder2007,x="gdpPercap",y="lifeExp",color="continent",size="pop",size_max=60,hover_name="country",facet_col="continent",log_x=True)
#hover_name="country" -> Controls which column's values appear as the hover label when you mouse over a data point.
#facet_col -> Splits the plot into subplots (facets), each showing data for a different continent.
```

```python
px.scatter(gapminder, x="gdpPercap", y="lifeExp",size="pop", size_max=60, color="continent", hover_name="country", 
           animation_frame="year", animation_group="country", log_x=True, range_x=[100,100000], range_y=[25,90],
           labels=dict(pop="Population", gdpPercap="GDP per Capita", lifeExp="Life Expectancy"))
#interested in more than just 2007 and you want to see how this chart evolved over time.
#You can animate it by setting `animation_frame="year"` and `animation_group="country"` to identify which circles match which ones across frames.
```

```python
px.choropleth(gapminder,locations="iso_alpha",color="lifeExp",hover_name="country",animation_frame="year",
              color_continuous_scale=px.colors.sequential.Plasma,projection="natural earth")
# A choropleth map is a type of visualization where regions (countries, states, etc.) are colored or 
# shaded based on a numeric variable (e.g., life expectancy, GDP, population).
```

```python
tips=px.data.tips() 
#tips = px.data.tips() loads the "tips" dataset from Plotly Express's built-in sample datasets.
#look below-> total_bill,tip,smoker,size  etc.. are some of the  variables from dataset)
```

```python
px.histogram(tips,x="total_bill",y="tip",histfunc="sum",color="smoker")
# Histogram graph is generated
```

```python
px.box(tips,x="total_bill",y="day",orientation="h",color="smoker",notched=True,category_orders={"day":["Thur","Fri","sat","sun]"]})
# Creates a box plot
# notched -->adds a notch around the median, displaying its confidence interval.
```

```python
px.violin(tips,y="tip",x="smoker",color="sex",box=True,points="all")
# violin-> generates violin graph
# Splits the violins into two groups: smokers (Yes) vs. non-smokers (No).
# Box =True, Overlays a mini box plot inside the violin,
```

```python
px.scatter(tips,x="total_bill",y="tip",color="smoker",trendline="ols",marginal_x="violin",marginal_y="box")
# px.scatter() code, trendline="ols" adds an Ordinary Least Squares (OLS) regression line to the scatter plot. 
# minimizes the sum of squared errors between observed and predicted values.
```
